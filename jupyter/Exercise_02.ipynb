The following function generates two datasets.  You don't need to know how it works...

```python
import numpy as np

def generate_data ():
    N_samples = 100
    Sigma = [
        [ 1.0, 0.9 ],
        [ 0.9, 1.0 ]
    ]
    
    X_1 = np.random.default_rng().multivariate_normal([0.0, 0.0], Sigma, N_samples)
    X_2 = np.copy(X_1);
    X_2[:,0] = X_2[:,0]*1000 + 3000;
    X_2[:,1] = X_2[:,1] + 50;

    return (X_1, X_2)

X_1, X_2 = generate_data()
```

Let's take a look at a scatterplot of the first dataset.

```python
import matplotlib.pyplot as plt

ax_1 = plt.gca()
plt.scatter(X_1[:,0], X_1[:,1]);
plt.title('X_1')
ax_1.set_xlabel('f_1')
ax_1.set_ylabel('f_2');
```

**Are these two features highly correlated?**

Yes - given one feature value, you have a general idea of what the other fe4ature value will be.

Now let's take a look at the second dataset.

```python
ax_2 = plt.gca()
plt.scatter(X_2[:,0], X_2[:,1]);
plt.title('X_2')
ax_2.set_ylim(0, 500)
ax_2.set_xlim(0, 6000)
ax_2.set_xlabel('f_1')
ax_2.set_ylabel('f_2');
```

**Are these features highly correlated?**

It's kind of difficult to tell with this plot, since the y-axis range is so large but the f_2 values are fairly condensed around 50 or so. I think if we zoomed in the y-axis we would get a better idea. They could be correlated, or they could actually be in a horizontal line as they appear to be with these axis limits.

Let's see if seaborn's `pairplot` agrees.

```python
import seaborn as sns
import pandas as pd

sns.set(style="ticks", color_codes=True)

df_1 = pd.DataFrame(
    data=X_1,
    columns=['f_1', 'f_2']
)

df_2 = pd.DataFrame(
    data=X_2,
    columns=['f_1', 'f_2']
)

g_1 = sns.pairplot(df_1)
g_1.fig.suptitle('X_1', y=1.1);

g_2 = sns.pairplot(df_2)
g_2.fig.suptitle('X_2', y=1.1);
```

**According to the plots above, are the features strongly correlated in one dataset or both?**

My suspicions were correct! These features are strongly correlated in both datasets.

**Explain why the original plots didn't show all of the correlations.**

The axis limits in the 2nd plot made it appear as if there was no correlation, but it's just that the f_2 values all fell within a very small range (~48 to ~52), so stretching the f_2 axis from 0 to 500 forced all the values into a very small vertical portion of the plot.
